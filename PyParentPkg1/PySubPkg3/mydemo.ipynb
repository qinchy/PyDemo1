```python
stock_cnt = 200
trade_days = 504
stock_day_change = np.random.standard_normal((stock_cnt,trade_days))
stock_day_change
```

```python
import scipy.stats as scs
# 均值期望
stock_mean = stock_day_change[0].mean()
# 标准差
stock_std = stock_day_change[0].std()
print('股票0 mean均值期望:{:.3f}'.format(stock_mean))
print('股票0 std振幅标准差:{:.3f}'.format(stock_std))

# 绘制股票0的直方图
plt.hist(stock_day_change[0],bins=50)
# linspace从股票0最小值->最大值生成数据
fit_linspace = np.linspace(stock_day_change[0].min(), stock_day_change[0].max())

# 概率密度函数(pdf, probability density function)
# 由均值，方差来描述曲线，使用scipy.stats.norm.pdf生成拟合曲线
pdf = scs.norm(stock_mean, stock_std).pdf(fit_linspace)
# plot x,y
plt.plot(fit_linspace, pdf, lw=2, c='r')
```

```python
keep_days = 50
stock_day_change_test = stock_day_change[:stock_cnt,:trade_days-keep_days]
stock_day_change_test
```

```python
total_change = np.sum(stock_day_change_test, axis=1)
total_change
```

```python
stock_day_change_index = np.argsort(total_change)
stock_day_change_index[:3]
```

```python
def show_buy_lower(stock_index):
    """
    :param stock_index：股票序号，即再stock_day_change_data中的数据
    """
    # 设置一个一行两列的可视化图表
    _, axs = plt.subplots(nrows=1,ncols=2,figsize=(16,5))
    # view_days 504 - keep_days 50 = 454
    axs[0].plot(np.arange(0,trade_days - keep_days), stock_day_change_test[stock_index].cumsum())
    cs_buy =stock_day_change[stock_index][trade_days-keep_days : trade_days].cumsum()
    axs[1].plot(np.arange(trade_days - keep_days, trade_days),cs_buy)
    return cs_buy[-1]
```

```python
profit = 0.0
for stock_index in stock_day_change_index[:3]:
    p = show_buy_lower(stock_index)
    print('买入第{}支股票的收益为：{}'.format(stock_index,p))
    profit += p
print('买入第{}支股票，从第454个交易日开始持有盈亏：{:.2f}%'.format(stock_day_change_index[:3], profit))   
```

```python
# 设置100个赌徒
gamblers = 100

def casino(win_rate, win_once=1, loss_once=1, commission=0.01):
    """
        赌场：简单设定每个赌徒一共有1000000一共想在赌场玩10000000次，
        但是你要是没钱了也别想玩了
        win_rate:   输赢的概率
        win_once:   每次赢的钱数
        loss_once:  每次输的钱数
        commission: 手续费这里简单的设置了0.01 1%
    """
    my_money = 1000000
    play_cnt = 10000000
    commission = commission
    for _ in np.arange(0, play_cnt):
        # 使用伯努利分布根据win_rate来获取输赢
        w = np.random.binomial(1, win_rate)
        if w:
            # 赢了 +win_once
            my_money += win_once
        else:
            # 输了 -loss_once
            my_money -= loss_once
        # 手续费
        my_money -= commission
        if my_money <= 0:
            # 没钱就别玩了，不赊账
            break
    return my_money
```

```python
# 如果有numba使用numba进行加速, 这个加速效果非常明显，不使用numba非常非常非常慢
import numba as nb
casino = nb.jit(casino)
```

```python
# 假设骗子赌场有老千，有抽头的情况，
# 且win_once提高两个点，即win_once从1.0到1.02，
# 且loss_once降低两个点，即loss_once从1.0到0.98
# 综合盈利何亏损，得出赔率=1.02/0.98=1.040816
# 即100个赌徒进场赌场，胜率0.5，手续费0.01，赔率1.04
moneys = [casino(0.5, win_once=1.02, loss_once=0.98, commission=0.01) for _ in np.arange(0,gamblers)]
moneys
```
